In this quiz, you will walk through building a web scraper that pulls data from the internet into a pandas dataframe for analysis. The link that you will be using to practice is this URL: [https://btech-data-analytics.github.io/bridgerland-technical-college/employees_3.html](https://btech-data-analytics.github.io/bridgerland-technical-college/employees_3.html)

Go to the page shown above and observe the table of data on the page. In this project, you will follow along with the videos to scrape this data into a pandas dataframe. Afterwards, you will compute some summary statistics on the data set.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get(' https://btech-data-analytics.github.io/bridgerland-technical-college/employees_3.html').text
```

```python
soup = BeautifulSoup(response, 'html.parser')
```

```python
id = []
name = []
age = []

for employee in soup.find('table').find_all('tr', class_ = 'employee'):
    id.append(employee.find_all('td')[0].text)
    name.append(employee.find_all('td')[1].text)
    age.append(employee.find_all('td')[2].text)
```

```python
print(id)
print(name)
print(age)
```

```python
df = pd.DataFrame({
    'id' : id,
    'name' : name,
    'age' : age
})
df.info()
```

```python
df['id'] = df['id'].astype('Int64')
df['age'] = df['age'].astype('Int64')
df.info()
```

### **Question 1**
What is the average age of employees found at the link above?
Group of answer choices
* 29.875
* 30.175
* 28.125
* 29.667
### Answer: **29.875**

```python
print(df['age'].mean())
```

### **Question 2**

Group of answer choices

### Answer: **3.56**
