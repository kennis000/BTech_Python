```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('companies.csv')
df.info()
df.head()
```

### **Question 1**
Using matplotlib/seaborn, create a countplot that shows a <u>count of how many of the companies in the data set are from each state</u>.

Paste a picture of the graph below. Note that you may need to adjust the width/height of the graph appropriately.


```python
plt.figure(figsize=(13, 4))
sns.countplot(x='state', data = df.sort_values(by = 'state'), hue = 'state', legend = False, palette = 'flare')
plt.show()
```

### **Question 2**
Which state are most of the companies from?
Group of answer choices
* California (CA)
* Texas (TX)
* New York (NY)
* Utah (UT)

### Answer: <u>California (CA)</u>

### **Question 3**

Using matplotlib/seaborn, create a bar chart that shows <u>a sum of how many total job openings there are in each industry</u>.

Paste a picture of the graph below. Note that you may need to adjust the width/height of the graph appropriately.

```python
plt.figure(figsize=(13, 6))
sns.barplot(y = 'Industry', x = 'job_openings', data = df[['Industry', 'job_openings']].groupby('Industry').sum().sort_values('Industry'), hue = 'Industry', legend = False, palette = 'pastel')
plt.show()
```

### **Question 4**
Which industry has the most job openings?
Group of answer choices
* Security
* Analytics
* Fintech
* Finance

### Answer: <u>Fintech</u>

### **Question 5**

Using matplotlib/seaborn, create a <u>line chart that shows the average growth rate for each founded year</u>.

Paste a picture of the graph below. Note that you may need to adjust the width/height of the graph appropriately.

```python
plt.figure(figsize=(10, 4))
sns.lineplot(x = 'founded', y = 'growth_percentage', data = df[['founded','growth_percentage']].groupby('founded').mean(), color = 'red').set(ylabel = 'Avg_Growth_%', xlabel = 'Founding Year')
plt.show()
```

### **Question 6**
From the years 2000 to 2020, does the growth percentage seem to be increasing, decreasing, or not changing at all?
Group of answer choices
* increasing
* decreasing
* no change at all

### Answer: <u>increasing</u>

### **Question 7**

Using matplotlib/seaborn, create a <u>scatter plot that compares the number of employees to estimated revenues for each company</u>.

Paste a picture of the graph below. Note that you may need to adjust the width/height of the graph appropriately.

```python
plt.figure(figsize=(12, 4))
sns.scatterplot(x = 'employees', y = 'estimated_revenues', data = df[['employees','estimated_revenues']], alpha = 0.6)
plt.show()
```
